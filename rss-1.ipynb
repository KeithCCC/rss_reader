```python
import feedparser
```

```python
import extractrss
```

```python
#Source site url
source_web_site = "https://atmarkit.itmedia.co.jp/"

# URL of the RSS feed
rss_url = "https://rss.itmedia.co.jp/rss/2.0/ait.xml"
```

```python
collection = extractrss.extract_rss_urls(source_web_site)
```

```python
for item in collection:
    feed = feedparser.parse(item)
    for entry in feed.entries:
        if 'published' in entry:
            print(entry.published)
        print(entry.title)
        print(entry.link)
        print(entry.summary)
        print('---------------------------------')
        print()
```
